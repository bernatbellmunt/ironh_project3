```python
def get_restaurants(query,latitude,longitude):

    url = f"https://api.foursquare.com/v3/places/nearby?ll={latitude}%2C{longitude}&query={query}"

    headers = {
        "accept": "application/json",
        "Authorization": f"{key}"}
    response = requests.get(url, headers=headers).json()
    
    return response


get_restaurants("Restaurant",41.397974881307356, 2.190445368837462)
```

```python
queryicon = Icon (
        color="blue",
        opacity = 0.6,
        prefix = "fa",
        icon="sun-o",
        color_icon = "black"
)

def place_in_map(query,latitude,longitude):

    url = f"https://api.foursquare.com/v3/places/nearby?ll={latitude}%2C{longitude}&query={query}"

    headers = {
        "accept": "application/json",
        "Authorization": f"{key}"}
    response = requests.get(url, headers=headers).json()

    newlist=[]
    for lista in response["results"]:
        lat= lista["geocodes"]["main"]["latitude"]
        long = lista["geocodes"]["main"]["longitude"]
        name = lista["name"]
        address= lista["location"]["address"]
        
        cat_list
        for d in lista["categories"]:
            cat_list.append(d["name"])
        
        categories = cat_list
        newlist.append({"name":name,"lat":lat,"long":long, "categories":categories, "address":address})
    
    df = pd.DataFrame(newlist)
    map_3 = Map(location = [latitude,longitude], zoom_start=15)
    
    for index,row in df.iterrows():

        #primer iconi, depres marcador
        icono = {"location":[row["lat"],row["long"]]}

        marker = Marker(**icono)
    
    
    
        
    marker.add_to(map_3)    
    return map_3
        
    
place_in_map("bar",41.397974881307356, 2.190445368837462)
```

```python
def get_dataframe(query,latitude,longitude):

    url = f"https://api.foursquare.com/v3/places/nearby?ll={latitude}%2C{longitude}&query={query}"

    headers = {
        "accept": "application/json",
        "Authorization": f"{key}"}
    response = requests.get(url, headers=headers).json()

    newlist=[]
    for lista in response["results"]:
        lat= lista["geocodes"]["main"]["latitude"]
        long = lista["geocodes"]["main"]["longitude"]
        name = lista["name"]
        address= lista["location"]["address"]
        newlist.append({"name":name,"lat":lat,"long":long, "address":address})
    
    df = pd.DataFrame(newlist)
    return df

df = get_dataframe("bar",41.397974881307356, 2.190445368837462)
```
