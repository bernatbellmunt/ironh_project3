```python
import os
key = os.getenv("4squarekey")
#key
```

```python
def get_restaurants(query,latitude,longitude):

    url = f"https://api.foursquare.com/v3/places/nearby?ll={latitude}%2C{longitude}&query={query}"

    headers = {
        "accept": "application/json",
        "Authorization": f"{key}"}
    response = requests.get(url, headers=headers).json()
    
    return response


get_restaurants("Restaurant",41.397974881307356, 2.190445368837462)
```

```python
restaurants_close_to = get_restaurants("Restaurant",41.397974881307356, 2.190445368837462)

restaurants_close_to
```

```python
restaurants_close_to["results"]
```

```python
restaurants_close_to["results"][0].keys()
```

```python
iconrestaurant = Icon (
        color="blue",
        opacity = 0.6,
        prefix = "fa",
        icon="sun-o",
        color_icon = "black"
)
```

```python
map_2 = Map(location = [41.397974881307356, 2.190445368837462], zoom_start=15)
for lista in restaurants_close_to["results"]:
    
    lat= lista["geocodes"]["main"]["latitude"]
    long = lista["geocodes"]["main"]["longitude"]
    marcador = Marker(
    location = [lat, long],
    tooltip = lista["name"],
    icon=iconrestaurant)
        
    marcador.add_to(map_2)
        
map_2
```

```python
newlist=[]
for lista in restaurants_close_to["results"]:
    lat= lista["geocodes"]["main"]["latitude"]
    long = lista["geocodes"]["main"]["longitude"]
    name = lista["name"]
    newlist.append({"name":name,"lat":lat,"lon":long})
df = pd.DataFrame(newlist)
df
```

```python
map_2 = Map(location = [41.397974881307356, 2.190445368837462], zoom_start=15)
for index,row in df.iterrows():
    
    #primer iconi, depres marcador
    icono = {"location":[row["lat"],row["lon"]]}
    
    marker = Marker(**icono)
    
    
    
        
    marker.add_to(map_2)
        
map_2
```
